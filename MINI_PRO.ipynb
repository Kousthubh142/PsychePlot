```python
import os
from groq import Groq
from dotenv import load_dotenv
from IPython.display import display, Markdown
```

```python
load_dotenv()
client=Groq(api_key=os.getenv("groq_api_key"))
```

```python
story_history=[]
def answerit(query: str)->str:
  prompt = f"""
    Generate the continuation of the story which starts like :
    - Start of the story : It's 11:00 pm on a winter night, you are standing next to a phone booth, you are witnessing a murder scene, and then the murderers see you standing behind the booth and watching the scene.
    
    Step by step process : 
    1. Analyze the query
    2. Give contuation story of the query
    3. Provide 4 options(emotional, reasonal, diplomatic, angry types) which can be the continuation of the story further 

    - Follow this templete : 
    1. Analysed the query : [detailed understanding]
    2. Story : [Generated story not more than 150 words]
    3. Options : [4 options numbered 1,2,3,4 without the order of the types]
  """
  response = client.chat.completions.create(
      messages=[{"role": "user", "content": prompt}],
      model="llama3-70b-8192", 
      temperature=0.3,
      max_tokens=600
  )
  story_segment= response.choices[0].message.content
  story_history.append(story_segment)
  return story_segment

def continue_story(selected_option: str) -> str:
    full_story = "\n".join(story_history)  
    
    prompt = f"""
    Continue the following story based on the user's choice:

    {full_story}

    Selected Option: "{selected_option}"

    - Continue the story while maintaining consistency.
    - Generate another four options for the user to choose from.
    """
    
    response = client.chat.completions.create(
        model="llama3-70b-8192",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.3,
        max_tokens=600
    )
    
    story_segment =  response.choices[0].message.content
    story_history.append(story_segment)  
    return story_segment

display(Markdown("# PSYCHEPLOT"))
print("\n\n")
print("-"*200)
print("\n\n")
print(answerit("start"))
n=3
while (n>0):
    user_choice = input("Give me the number: ")
    continuation = continue_story(user_choice)
    print(continuation)
    n-=1  
```
